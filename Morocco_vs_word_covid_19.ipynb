La COVID-19 affecte les individus de différentes manières. La plupart des personnes infectées développent une forme légère à modérée de la maladie et guérissent sans hospitalisation.
Symptômes les plus fréquents:
  fièvre
  toux sèche
  fatigue
Symptômes moins fréquents:
  courbatures
  maux de gorge
  diarrhée
  conjonctivite
  maux de tête
  perte de l’odorat ou du goût
  éruption cutanée, ou décoloration des doigts ou des orteils

# Nombre Total des cas 

```
#Total Active  represente  Total cases - (death + recovred)
df['Total Active'] = df['Confirmed'] - (df['Deaths'] + df['Recovered'])
total_active = df['Total Active'].sum()
print('Total number of active COVID 2019 cases across Morocco:', total_active)

total_active.style.background_gradient(cmap='Reds')
```
